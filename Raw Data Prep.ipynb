## Create csv file for unit_type table

### Attributes = flat_type {int} , rooms {string}

```python
#get all unique value counts of flat_type
room_types = sampled_df['flat_type'].unique()
room_types.sort()
flat_type = list(range(0,len(room_types)))
```

```python
unit_type_df = pd.DataFrame(data={'flat_type_id':flat_type,'flat_type':room_types})
```

```python
unit_type_df
```

```python
unit_type_df.info()
```

```python
#exported as csv
unit_type_df.to_csv("clean_dataset/unit_type.csv", index=False)
```

## Create csv file for sale_info table

### Attributes:
#### - info
#### - id
#### - month
#### - floor_area
#### - storey_id
#### - flat_type
#### - unit_model
#### - location_id


```python
#merge locations(town)
df_merged = pd.merge(sampled_df, locations_df, how='left',on='town')
```

```python
#merge flat type(rooms)
df_merged = pd.merge(df_merged, unit_type_df, how='left',on='flat_type')
```

```python
#merge flat model(Improved, Model A, etc)
df_merged = pd.merge(df_merged, unit_model_df, how='left',on='flat_model')
```

```python
#merge storey
df_merged = pd.merge(df_merged, storey_df, how='left',on='storey_range')
```

```python
sales_info_df = df_merged.drop(['street_name', 'lease_commence_date','block','resale_price','town','flat_type','storey_range','flat_model'], axis=1)
```

```python
sales_info_df.head()
```

```python
sales_info_df.insert(0,"id",sales_list)
```

```python
sales_info_df.head()
```

```python
#get the years only from lease, then convert to int
sales_info_df['lease_left'] = sales_info_df['remaining_lease'].str.split(" ").str[0]
sales_info_df['lease_left'] = sales_info_df['lease_left'].astype('int64')
```

```python
sales_info_df = sales_info_df[['id','date_sold','floor_area_sqm','lease_left','location_id','flat_type_id',
                               'model_id','storey_id','remaining_lease']]

sales_info_df = sales_info_df.drop(['remaining_lease'],axis=1)
sales_info_df = sales_info_df.rename(columns={'id':'sale_info_id'})
```
